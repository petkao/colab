```python
def isAnagram(s, t):
  if (len(s) != len(t)):
    return False
  for i in range(len(s)):
    if (s.count(s[i]) != t.count(s[i])):
      return False
  return True
```

```python
s = "racecar"
t = "carrace"
isAnagram(s, t)
```

```python
s = "jar"
t = "jam"
isAnagram(s, t)
```

```python
from collections import defaultdict

def twoSum(nums, target):
  prevMap = {} #val : index

  for i, n in enumerate(nums):
    diff = target - n
    if diff in prevMap:
      return [prevMap[diff], i]
    prevMap[n] = i
  return

```

```python
nums = [2,7,1,15]
target = 9
twoSum(nums, target)
```

```python
nums = [4,5,6]
target = 10
twoSum(nums, target)
```

```python
nums = [5,5]
target = 10
twoSum(nums, target)
```

```python
def groupAnagrams(strs):
  res = defaultdict(list)
  for s in strs:
    sortedS = "".join(sorted(s))
    res[sortedS].append(s)
  return res.values()
```

```python
strs = ["act","pots","tops","cat","stop","hat"]
groupAnagrams(strs)
```
