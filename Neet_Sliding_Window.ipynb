```python
def closeDuplicateBruteForce(nums, k):
  for L in range(len(nums)):
    for R in range(L + 1, min(len(nums), k + L)):
      if nums[L] == nums[R]:
        return True
  return False
```

```python
nums = [1,2,3,2,3,3]
k = 2
closeDuplicateBruteForce(nums, k)
```

```python
def closeDuplicate(nums, k):
  window = set()
  L = 0

  for R in range(len(nums)):
    print("R: ", R)
    print("L: ", L)
    print("nums[R]: ", nums[R])
    if R - L + 1 > k:
      window.remove(nums[L])
      L += 1
    if nums[R] in window:
      return True
    window.add(nums[R])
    print("window: ", window)
  return False

```

```python
nums = [1,2,3,2,3,3]
k = 2
closeDuplicate(nums, k)
```

```python
def shortestSubarray(nums, target):
  L = 0
  total = 0
  length = float("inf")

  for R in range(len(nums)):
    total += nums[R]
    while total >= target :
      length = min(length, R - L + 1)
      L += 1
      total -= nums[L]
  return length if length != float("inf") else -1
```

```python
nums = [2,3,1,2,4,3]
target = 6
shortestSubarray(nums, target)
```

```python
def maxProfit(prices):
  L = 0
  max_profit = 0

  for R in range(len(prices)):
    if prices[L] > prices[R]:
      L = R
    max_profit = max(max_profit, prices[R] - prices[L])
  return max_profit
```

```python
prices = [10,1,5,6,7,1]
maxProfit(prices)
```

```python
def lengthOfLongestSubstring(s):
  longest = 0
  L =0

  for R in range(len(s)):
    # print("R: ", R)
    # print("L: ", L)
    # print("s[R]: ", s[R])
    # print("s[L:R]: ", s[L:R])
    if s[R] in s[L:R]:
      # print("I am here. ")
      # print("s[L:R]: ", s[L:R])
      index_in_window = s[L:R].find(s[R])
      if index_in_window != -1:
        L += index_in_window + 1

      # print("L: ", L)
      # print("s[L:R}.find(s[R]): ", index_in_window)
    longest = max(longest, R - L + 1)
  return longest
```

```python
def characterReplacement(s, k):
    count = {}
    max_length = 0

    l = 0
    max_freq = 0
    for r in range(len(s)):

        print("count before: ", count)
        print("r: ", r)
        print("l: ", l)
        print("s[r]: ", s[r])
        print("s[l]: ", s[l])
        count[s[r]] = 1 + count.get(s[r], 0)
        max_freq = max(max_freq, count[s[r]])

        print("max_freq: ", max_freq)

        while (r - l + 1) - max_freq > k:
            count[s[l]] -= 1
            l += 1
        max_length = max(max_length, r - l + 1)
        print("count after: ", count)

    return max_length
```
