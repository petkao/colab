```python
def removeDuplicates(nums) -> int:
    nums.sort()
    j = 1
    for i in range(1, len(nums)):
        if nums[i] != nums[i - 1]:
            nums[j] = nums[i]
            j += 1
    return j
```

```python
nums = [1,1,2]

removeDuplicates(nums)
```

```python
def maxProfit(prices) -> int:
  i = 0
  j = 0
  profit = 0
  diff = 0
  price_diff = [0] * len(prices)
  for i in range(0, len(prices)):
    j = i
    # print("i: ",i)

    for j in range(i + 1, len(prices)):
      # print("j: ",j)
      diff = prices[j] - prices[i]
      # print("diff: ", diff)
      # print("price_diff[j]: ",price_diff[j])

      if diff >= price_diff[i]:
        price_diff[i] = diff
      # print("price_diff: ",price_diff)

  for i in range(0, len(prices)):

    if price_diff[i] > profit:
      profit = price_diff[i]

  print("profit: ",profit)

  return profit
```
