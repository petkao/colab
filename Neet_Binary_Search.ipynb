```python
def search(nums, target):
  l = 0
  r = len(nums) - 1

  while l <= r:
    mid = (l + r) // 2
    if nums[mid] == target:
      return mid
    elif nums[mid] < target:
      l = mid + 1
    else:
      r = mid - 1
  return -1
```

```python
nums = [-1,0,2,4,6,8]
target = 4
search(nums, target)
```

```python
def findMin(nums):
  res = nums[0]
  l = 0
  r = len(nums) - 1
  # print("nums: ", nums)

  while l <= r:
    # print("r: ", r)
    # print("l: ", l)
    if nums[l] < nums[r]:
      res = min(res, nums[l])
      break

    mid = (l + r) // 2
    res = min(res, nums[mid])
    if nums[mid] >= nums[l]:
      l = mid + 1
    else:
      r = mid - 1
  return res

```

```python
nums = [3,4,5,6,1,2]
findMin(nums)
```

```python
nums = [4,5,0,1,2,3]
findMin(nums)
```

```python
nums = [4,5,6,7]
findMin(nums)
```

```python
def searchRotatedSortedArray(nums, target):

  l = 0
  r = len(nums) - 1

  while l <= r:
    mid = (l + r) //2

    if nums[mid] == target:
      return mid

    # left sorted portion
    if nums[l] <= nums[mid]:
      if target > nums[mid] or target < nums[l]:
        l = mid + 1
      else:
        r = mid - 1

    # right sorted portion
    else:
      if target < nums[mid] or target > nums[r]:
        r = mid - 1
      else:
        l = mid + 1

  return -1

```

```python
nums = [3,4,5,6,1,2]
target = 1
searchRotatedSortedArray(nums, target)
```

```python
nums = [3,5,6,0,1,2]
target = 4
searchRotatedSortedArray(nums, target)
```
