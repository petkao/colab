```python
def minSubArrayLen(target, nums):
  i = 0
  # array size
  j = 1
  test = 0
  nums.sort(reverse = True)
  print(nums)
  while i < len(nums):
    print("i: ", i)
    print("test: ",test)
    test += nums[i]
    if test >= target:
      return j
    else:
      j += 1
    i += 1

  return 0




```

```python
nums= [2,3,1,2,4,3]
minSubArrayLen(7,nums)
```
