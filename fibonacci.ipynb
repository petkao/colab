```python
from collections import deque

def generate_fibonacci_up_to_n(n):
    fibs = [0, 1]
    while fibs[-1] <= n:
        fibs.append(fibs[-1] + fibs[-2])
    return fibs[2:]  # We skip the first two Fibonacci numbers: 0 and 1

def solution(A):
    n = len(A)
    fibs = generate_fibonacci_up_to_n(n + 1)
    print(fibs)
    queue = deque([(-1, 0)])  # (current_position, number_of_jumps)
    # print(queue)
    visited = set([-1])

    while queue:
        curr_pos, jumps = queue.popleft()
        for fib in fibs:
            next_pos = curr_pos + fib
            if next_pos == n:
                return jumps + 1
            if 0 <= next_pos < n and A[next_pos] == 1 and next_pos not in visited:
                visited.add(next_pos)
                print("visited: ", visited)
                queue.append((next_pos, jumps + 1))
                print(queue)

    return -1

A = [0, 0, 0, 1, 1, 0, 1, 0, 0, 0, 0]
solution(A)
```
