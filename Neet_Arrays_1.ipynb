```python
def hasDuplicate(nums):
  if set(nums) != len(nums):
    return True
  else:
    return False
```

```python
nums = [1, 2, 3, 3]
hasDuplicate(nums)
```

```python
def isAnagram_1(s, t):
  if len(s) != len(t):
    return False

  sortedS = sorted(s)
  sortedT = sorted(t)

  for i in range(len(s)):
    if sortedS[i] != sortedT[i]:
      return False

  return True
```

```python
def twoSum(nums, target):
  i = 0
  while i < len(nums):
    j = i + 1
    while j < len(nums):
      if nums[i] + nums[j] == target:
        return [i, j]
      else:
        j += 1
    i += 1

```

```python
def twoSum_1(nums, target):
  prevMap = {}
  for i, n in enumerate(nums):
    diff = target - n
    if diff in prevMap:
      return [prevMap[diff], i]
    prevMap[n] = i
  return

```

```python
nums = [3,4,5,6]
target = 7
twoSum_1(nums, target)
```

```python
nums = [4, 5, 6]
target = 10
twoSum(nums, target)
```

```python
from collections import defaultdict

students = [
    {'name': 'John', 'grade': 'A'},
    {'name': 'Alice', 'grade': 'A'},
    {'name': 'Bob', 'grade': 'B'},
    {'name': 'Charlie', 'grade': 'B'},
]

grade_groups = defaultdict(list)
for student in students:
    print("student: ", student)
    grade_groups[student['grade']].append(student['name'])

print(grade_groups)  # Output: defaultdict(<class 'list'>, {'A': ['John', 'Alice'], 'B': ['Bob', 'Charlie']})
```

```python
def test(strs):
  res = defaultdict(list)
  for s in strs:
    sortedS = "".join(sorted(s))
    res[sortedS].append(s)
  return list(res.values())
```

```python
strs = ["act","pots","tops","cat","stop","hat"]
test(strs)
```

```python
from collections import defaultdict

def groupAnagrams(strs):
  res = defaultdict(list)
  for s in strs:
    sortedS = "".join(sorted(s))
    res[sortedS].append(s)

  return list(res.values())

```

```python
from collections import defaultdict

def topKFrequent(nums, k):
  res = []
  i = 0
  counts = defaultdict(int)
  for x in nums:
    counts[x] += 1

  sorted_desc_values = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))

  for y in sorted_desc_values:
    if i < k:
      res.append(y)
    i += 1

  return res
```
