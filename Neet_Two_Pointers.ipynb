```python
def isPalindrome(s: str) -> bool:
    nospace_s = s.replace(" ","")
    lower_s = nospace_s.lower()
    new_s = ""
    for char in lower_s:
        if (ord(char) >=97 and ord(char) <= 122):
            # print("char: ", char)
            new_s = new_s + char
        if (ord(char) >=48 and ord(char) <= 57):
            # print("char: ", char)
            new_s = new_s + char
    # print("lower_s: ", lower_s)
    # print("new_s: ", new_s)
    reverse_s = new_s[::-1]
    # print("reverse_s: ", reverse_s)
    if new_s == reverse_s:
        return True
    else:
        return False
```

```python
s = "Was it a car or a cat I saw?"
isPalindrome(s)
```

```python
s = "tab a cat"
isPalindrome(s)
```

```python
def twoSum(numbers, target):
    i = 0
    result = []
    while i < len(numbers):
        j = i + 1
        while j < len(numbers):
            if (numbers[i] + numbers[j]) == target:
                print("numbers[i]: ", numbers[i])
                print("numbers[j]: ", numbers[j])
                result.append(numbers[i])
                result.append(numbers[j])
                return result
            j += 1
        i +=1
```

```python
numbers=[2,3,4]
target=6
twoSum(numbers,target)
```

```python
from collections import defaultdict
def twoSumPlus(numbers, target):
  mp = defaultdict(int)
  for i in range(len(numbers)):
    tmp = target - numbers[i]
    print("tmp: ",tmp)
    print("mp[tmp]: ", mp[tmp])
    print("i: ", i)
    if mp[tmp]:
      return [mp[tmp], i + 1]
    mp[numbers[i]] = i + 1
    print("mp: ",mp)
  return []
```

```python
numbers=[2,3,4]
target=6
twoSumPlus(numbers,target)
```

```python
def threeSum(nums):
    res = set()
    nums.sort()
    # print("nums: ",nums)
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            for k in range(j + 1, len(nums)):
                if nums[i] + nums[j] + nums[k] == 0:
                    tmp = [nums[i], nums[j], nums[k]]
                    res.add(tuple(tmp))
    return list(res)
```

```python
nums = [-1,0,1,2,-1,-4]
threeSum(nums)
```

```python
def maxArea(heights):
    res = 0
    for i in range(len(heights)):
        for j in range(i + 1, len(heights)):
            res = max(res, min(heights[i], heights[j]) * (j - i))
    return res
```

```python
heights = [1,7,2,5,4,7,3,6]
maxArea(heights)
```
