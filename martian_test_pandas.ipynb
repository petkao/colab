```python
# selecting rows based on condition
rslt_df_true = df[(df['openai/chat/gpt-3.5-turbo'] == True)]
rslt_df_false = df[(df['openai/chat/gpt-3.5-turbo'] == False)]
rslt_df_merge = pd.merge(rslt_df_true,rslt_df_false)
# print('\nResult dataframe :\n', rslt_df)
gpt_true_count = rslt_df_true['openai/chat/gpt-3.5-turbo'].count()
print(gpt_true_count)
gpt_false_count = rslt_df_false['openai/chat/gpt-3.5-turbo'].count()
print(gpt_false_count)
accuracy = gpt_true_count / (gpt_true_count + gpt_false_count)
print('openai/chat/gpt-3.5-turbo accuracy: ', accuracy)
```
