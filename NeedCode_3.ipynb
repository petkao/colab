```python
def hasDuplicate(nums):
  if len(nums) == 0:
    return False
  nums.sort()
  i = 1
  prev = nums[0]
  while i < len(nums):
      if nums[i] == prev:
          return True
      else:
          prev = nums[i]
      i += 1
  return False

nums = [1, 2, 3, 4]
print(hasDuplicate(nums))

nums = []
print(hasDuplicate(nums))
```

```python
def threeSum(nums):
  res = set()
  i = 0
  j = 0
  k = 0
  nums.sort()
  for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
      for k in range(j + 1, len(nums)):
        if nums[i] + nums[j] + nums[k] == 0:
          res.add(tuple((nums[i], nums[j], nums[k])))
  return list(res)

nums = [-1,0,1,2,-1,-4]
print(threeSum(nums))
```

```python
nums = [0,1,1]
print(threeSum(nums))
```

```python
def Threesum(nums):
  nums.sort()
  counts = defaultdict(int)
  for num in nums:
    counts[num] += 1

  res = []
  for i in range(len(nums)):
    counts[nums[i]] -= 1
    if i and nums[i] == nums[i - 1]:
      continue
    for j in range(i + 1, len(nums)):
      counts[nums[j]] -= 1
      if j != i + 1 and nums[j] == nums[j - 1]:
        continue
      complement = -nums[i] - nums[j]
      if complement in counts and counts[complement] > 0:
        res.append([nums[i], nums[j], complement])

    for j in range(i + 1, len(nums)):
      counts[nums[j]] += 1
    # counts[nums[i]] += 1
  return res

nums = [-1,0,1,2,-1,-4]
print(Threesum(nums))


```

```python
def trap(height) -> int:
    prefix = [0] * len(height)
    suffix = [0] * len(height)
    water = [0] * len(height)
    total_water = 0
    i = 0
    j = 0
    for i in range(len(height)):
        print("i:", i)
        prefix_val = 0
        suffix_val = 0
        for j in range(i):
            if height[j] > prefix_val:
                prefix_val = height[j]
        # prefix.append(prefix_val)
        prefix[i] = prefix_val
        print("prefix_val: ", prefix_val)
        k = len(height) - 1
        while k > i:
            if height[k] > suffix_val:
                suffix_val = height[k]
            # print("k: ", k)
            k -= 1

        print("suffix_val: ", suffix_val)
        suffix[i] = suffix_val

    for i in range(len(height)):
      if min(prefix[i], suffix[i]) - height[i] > 0:
        water[i] = min(prefix[i], suffix[i]) - height[i]

    for i in range(len(water)):
      total_water += water[i]

    return total_water
```
