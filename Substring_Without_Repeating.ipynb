```python
def lengthOfLongestSubstring(s):
  i = 0

  subs = s[0]
  maxsubs = s[0]
  while i < len(s) - 1:
    print("i: ",i)
    print("s[i]: ", s[i], "s[i+1]: ", s[i+1])
    print("subs: ", subs)
    if s[i] != s[i+1] and s[i+1] not in subs:
      subs = subs + s[i+1]
      print("subs: ", subs)
      i += 1
    else:
      print("maxsubs: ", maxsubs)
      print("subs: ", subs)
      if len(subs) >= len(maxsubs):
        maxsubs = subs
        print("maxsubs: ", maxsubs)
      subs = s[i]
      i += 1
  return len(maxsubs)

```

```python
s = "abcabcbb"
lengthOfLongestSubstring(s)
```
